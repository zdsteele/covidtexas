```python
cols = []
raw_cols = list(df.columns)
```

```python
for col in raw_cols:
    
    try:
        
        new = col.replace('\n', '')
        cols.append(new)
        
    except:
        
        cols.append(col)
```

```python
res = dict(zip(raw_cols, cols))
```

```python
df = df.rename(columns=res)
```
